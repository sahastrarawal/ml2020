```
#in case of using lap/pc
#pip install matplotlib seaborn


x=[2,5,4,9]
y=[3,8,3,6]
```

```
import matplotlib.pyplot as pit
```

```
pit.xlabel('distance')#for x
pit.ylabel('time')#for y
pit.scatter(x,y,label='missles')# to show dots only
pit.scatter(y,x,label='target')
pit.plot(x,y) #for joing dots by line
pit.bar(x,y,color='red') #bar plot
pit.bar(y,x,color='yellow')
pit.grid(c='cyan')#or grid and colour as well
pit.legend()  #to show labels of plot area
pit.plot()# to plot graph
```

```
#cricket score graph
players=["dhoni","rohit","virat"]
runs=[120,160,130]
```

```
pit.xlabel("players")
pit.ylabel("runs")
#pit.bar(players,runs,label="2020 starting match")
pit.pie(runs,labels=players,autopct='%1.1f%%',explode=(0,0.3,0),shadow=True)
pit.legend()
pit.show()
```
